```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
# 회귀분석 데이터 셋 
# 데이터셋 pandas를 통해 데이터 셋 불러오기
# 데이터셋 구성 : 
# 자치구별 / 인구수 : 흡연구역개수 : 단속건수 : 유입인구 : 흡연률 : CCTV : 강우량  - > X 값 독립변수
# 침수피해 -> Y 값 종속변수
df = pd.read_excel("./data/회귀분석데이터셋.xlsx")
df
```

```python
df.columns
```

```python
# X값과Y값으로 나눔
x = df[['인구수','흡연구역개수','단속건수','유입인구','흡연률','CCTV','강우량']]
y=  df['침수피해']
```

```python
x.feature = x.columns 

from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler를 통해 scale을 맞춰줌
# MinMaxScaler객체 생성
scaler = MinMaxScaler()

# MinMaxScaler 로 데이터 셋 변환. fit() 과 transform() 호출.
scaler.fit(x)
data_scaled = scaler.transform(x)

# transform()시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
x_scaled = pd.DataFrame(data = data_scaled, columns=x.feature)

x_scaled.index = y.index # 인덱스가 달라서 똑같이 설정

x_scaled
```

```python
import statsmodels.api as sm
feature_add = sm.add_constant(x_scaled, has_constant='add')

# OLS model을 통해 각각의 독립변수(x)가 종속변수(y)에 영향이 있는지 확인
# p값이 일반 알파 수준인 0.05 보다 큰 값들은 통계적으로 유의하지 않다고 본다.
# sm OLS 적합
model = sm.OLS(y , feature_add)
fitted_model = model.fit()

# summary 함수통해 결과출력
fitted_model.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
X_train = x_scaled

def feature_engineering_XbyVIF(X_train):
    vif = pd.DataFrame()
    vif["VIF_Factor"] = [variance_inflation_factor(X_train.values,i)
                         for i in range(X_train.shape[1])]
    vif["Feature"] = X_train.columns
    return vif
vif = feature_engineering_XbyVIF(X_train)
print(vif)
```
